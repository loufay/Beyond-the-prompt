```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import base64
```

```python
PATH_TO_DATA = "/mnt/data2/datasets_lfay/MedImageInsights/data"
read_path = os.path.join(PATH_TO_DATA, "CheXpert-v1.0-512")

df_train = pd.read_csv(os.path.join(read_path, "train.csv"))
```

```python
def read_image(image_path):
    with open(image_path, "rb") as f:
        return f.read()
```

```python
## extract three rows where Pneumonia is 1 and race is White
df_train_pneumonia = df_train[df_train.Pneumonia == 1]
df_train_pneumonia_white = df_train_pneumonia[df_train_pneumonia.race == "White"]
df_train_pneumonia_white = df_train_pneumonia_white.sample(3)

df_train_pneumonia_asian = df_train_pneumonia[df_train_pneumonia.race == "Asian"]
df_train_pneumonia_asian = df_train_pneumonia_asian.sample(3)

df_train_pneumonia_black = df_train_pneumonia[df_train_pneumonia.race == "Black"]
df_train_pneumonia_black = df_train_pneumonia_black.sample(3)
```

```python
# extract path of first entry in df_train_pneumonia_white
image_path = os.path.join(read_path, df_train_pneumonia_white.iloc[0].Path)
image_path
```

```python
# plot white
images_white = [
    PATH_TO_DATA + row["Path"]
    for _, row in df_train_pneumonia_white.iterrows()
]

images_asian = [ 
    PATH_TO_DATA + row["Path"]
    for _, row in df_train_pneumonia_asian.iterrows()
]

images_black = [
    PATH_TO_DATA + row["Path"]
    for _, row in df_train_pneumonia_black.iterrows()
]
```

```python
# plot white images in one row, asian in another row, black in another row
fig, axs = plt.subplots(3, 3, figsize=(10, 10))
for i, image in enumerate(images_white):
    img = plt.imread(image)
    axs[0, i].imshow(img)
    axs[0, i].axis("off")
for i, image in enumerate(images_asian):
    img = plt.imread(image)
    axs[1, i].imshow(img)
    axs[1, i].axis("off")
for i, image in enumerate(images_black):
    img = plt.imread(image)
    axs[2, i].imshow(img)
    axs[2, i].axis("off")

```
